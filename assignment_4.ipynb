```
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
import numpy as np

data = pd.read_csv('google_review_ratings.csv')
#purify null data
print(data.shape)
simputer = SimpleImputer(missing_values=np.NAN, strategy='median')
data = data.iloc[:, 1:(data.shape[1])]
data_transformed = simputer.fit_transform(X=data.values)

print(data.shape)
```

Perform the clustering with DBSCAN and compute its Silhouette coefficient.

```
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import numpy

def Estimate_eps_n_min_pts(X, K):
        samples = X.shape[0]
        nn = NearestNeighbors(n_neighbors=(K + 1))
        neighbors = nn.fit(X)
        distances, indices = neighbors.kneighbors(X)
        distance_mtx = numpy.empty([K, samples])
        min_pts = []
        eps = []
        for i in range(0, K):
            dst = distances[:, (i+1)]
            dst.sort()
            dst = dst[::-1]
            lst_cost = numpy.inf
            for j in range(0, samples):
                cost = dst[j] + 0.05*(numpy.linalg.norm(range(0, j), ord=2))
                if lst_cost < cost:
                    min_pts.append(j-1)
                    eps.append(dst[j-1])
                    break
                lst_cost = cost
            distance_mtx[i] = dst

        min_pts = min_pts[numpy.argmin(eps)]
        eps = eps[numpy.argmin(eps)]

        fig = plt.figure()
        for i in range(0, K):
            plt.plot(distance_mtx[i])
        plt.ylabel('distance')
        plt.xlabel('points')
        plt.title('plot of estimate eps and min_pts')
        plt.axvline(x=min_pts, color='red', linestyle='--')
        plt.axhline(y=eps, color='red', linestyle='--')
        return min_pts, eps
def ComputeCenters(cluster, X):
        un_clust = numpy.unique(cluster)
        center = []
        for cl in un_clust:
            sub_X = X[sum(numpy.where(cluster == cl)), :]
            mn_pts = []
            for i in range(0, sub_X.shape[1]):
                mn_pts.append(numpy.mean(sub_X[:, i]))
            center.append(mn_pts)
        return center

def ComputeSSE(cluster, centers, X):

        un_clust = numpy.unique(cluster)
        cl_count = 0
        error_ov = []
        for cl in un_clust:
            sub_X = X[sum(numpy.where(cluster == cl)), :]
            error_clust = []
            for i in range(0, sub_X.shape[0]):
                error = numpy.subtract(sub_X[i,:], centers[cl_count])**2
                error = sum(error)
                error_clust.append(error)
            error_ov.append(sum(error_clust))
            cl_count = cl_count + 1
        return numpy.sum(error_ov)

def DBSCAN_run(X):
    # Record fitting runtime
    print("calculating eps and min_samples for DBSCAN...")
    min_pt_d, eps_d = Estimate_eps_n_min_pts(X=X, K=13)
    start_time = time.time()
    print("Clustering X using DBSCAN")
    dbscan = DBSCAN(eps=eps_d, min_samples=min_pt_d).fit(X)
    labels = dbscan.labels_
    #Computing "the Silhouette Score"
    print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(X, labels))
    centers = ComputeCenters(labels, X)
    sse = ComputeSSE(labels, centers, X)
    print(" DBSCAN SSE is: % 0.2f" % sse)
    endTime = time.time() - start_time
    print("Clustering ended in " + "{0:.2f}".format(round(endTime, 2)) + " seconds")
    fig, axs = plt.subplots(2)
    fig.suptitle('below is kmeans cluster plot')
    axs[1].scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
    plt.show()
DBSCAN_run(data_transformed)
```
